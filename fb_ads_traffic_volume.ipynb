```python
import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh import palettes

TOOLS = 'crosshair,save,pan,box_zoom,reset,wheel_zoom,hover'
p = figure(x_axis_type="datetime", title="Age groups", tools = TOOLS)


```

```python
one = age.loc[age['age']=='13-17', ]
two = age.loc[age['age']=='18-24', ]
three = age.loc[age['age']=='25-34', ]
four = age.loc[age['age']=='35-44', ]
five = age.loc[age['age']=='45-54', ]
six = age.loc[age['age']=='55-64', ]
seven = age.loc[age['age']=='65+', ]
```

```python
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Reach'

p.line(one['date'], one['reach'], color=palettes.Category20[7][0], legend='13-17')
p.line(two['date'], two['reach'], color=palettes.Category20[7][1], legend='18-24')
p.line(three['date'], three['reach'], color=palettes.Category20[7][2], legend='25-34')
p.line(four['date'], four['reach'], color=palettes.Category20[7][3], legend='35-44')
p.line(five['date'], five['reach'], color=palettes.Category20[7][4], legend='45-54')
p.line(six['date'], six['reach'], color=palettes.Category20[7][5], legend='55-64')
p.line(seven['date'], seven['reach'], color=palettes.Category20[7][6], legend='65+')
p.legend.location = "top_left"
show(p)
```
